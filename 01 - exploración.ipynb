```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Gráficos
%matplotlib inline
sns.set(style="whitegrid")
```

```python
df = pd.read_csv('train.csv')

# Mostrar las primeras filas del dataset
print("Primeras filas del dataset:")
print(df.head())

```

```python
print("\nCantidad de estudiantes por programa académico:")
print(df['ESTU_PRGM_ACADEMICO'].value_counts().head(10))  # Solo los 10 más comunes

plt.figure(figsize=(10,5))
sns.countplot(y='ESTU_PRGM_ACADEMICO', data=df, order=df['ESTU_PRGM_ACADEMICO'].value_counts().iloc[:10].index, palette='mako')
plt.title('Top 10 Programas Académicos con más Estudiantes')
plt.xlabel('Cantidad de Estudiantes')
plt.ylabel('Programa Académico')
plt.tight_layout()
plt.show()
```

```python
print("\nDistribución por estrato de vivienda:")
print(df['FAMI_ESTRATOVIVIENDA'].value_counts())

plt.figure(figsize=(6,4))
sns.countplot(x='FAMI_ESTRATOVIVIENDA', data=df, palette='pastel', order=df['FAMI_ESTRATOVIVIENDA'].value_counts().index)
plt.title('Distribución por Estrato Socioeconómico')
plt.xlabel('Estrato')
plt.ylabel('Cantidad')
plt.tight_layout()
plt.show()
```
